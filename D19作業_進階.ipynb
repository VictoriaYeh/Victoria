```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('./boston.csv')
```

```python
#1.畫出箱型圖，並判斷哪個欄位的中位數在300~400之間?
df1.boxplot()
df2=df1.median().between(300,400)
df3=df2.to_frame()
print(df3[df3[0] == True])
```

```python
#2. 畫出散佈圖 x='NOX', y='DIS' ，並說明這兩欄位有什麼關係?
x=df1['NOX']
y=df1['DIS']
plt.scatter(x,y)
```
